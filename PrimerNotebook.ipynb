# Importar librerias
Es fundamental cuando trabajamos con Notebooks que especifiquemos librerias con las que trabajaremos, ya que si no lo hacemos gran parte del código nos dará error.

Si bien en muchos lenguajes de programación es común ponerlos al inicio del código, para este tipo de trabajo inicialmente es una buena práctica ponerlo en el bloque donde lo usaremos por primera vez para saber en donde usamos esa libreria.

## Pandas

```python
import pandas as pd
#Configuración de parametros de Pandas para mejor visualización
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.precision', 3)
pd.set_option('plotting.backend', 'matplotlib') 
pd.options.mode.chained_assignment = None

# En produccion eliminar, se incluye por informacion
print(f'Pandas Version: {pd.__version__}') 
```

## NumPy

```python
import numpy as np

np.set_printoptions(suppress=True)

# En produccion eliminar, se incluye por informacion
print(f'Numpy version: {np.__version__}')
```
