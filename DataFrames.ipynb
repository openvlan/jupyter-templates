```python
df_excel = df_excel.reset_index()
```

```python
pd.options.mode.chained_assignment = None  # default='warn'

#Corregir campo fecha
df_excel.created_at = df_excel.created_at.str.replace('T', '', regex=True) 
df_excel.created_at = df_excel.created_at.str.replace('.000Z', '', regex=True)

i = 0
for i in range(len(df_excel)): 
  #print(df[i])
  fecha=0
  hora=0
  fecha  = df_excel.created_at[i][:10]
  #print(fecha)
  df_excel.created_at[i] = fecha# + ' ' + hora
  i = i + 1 
```

```python
df_excel.head(1)
```

```python
df_excel['NuevoAtributo'] = 0
df_excel.NuevoAtributo = 'PRUEBA'
df_excel.head(2)
```

### Lower Casing

Pasar cadenas a minuscula

```python
df_excel['message_lower'] = 0
df_excel.message_lower = df_excel.message.str.lower()
df_excel.head(1)
```

```python
# Ayuda para expresiones regulares
# https://docs.python.org/es/3/howto/regex.html
df_excel.message_lower = df_excel.message_lower.str.replace(r'@[a-zA-z]+:?', '', regex=True) #Elimina @xxx 
#Anteriormente lo cambiamos por "@xxx"
df_excel.head(1)
```
